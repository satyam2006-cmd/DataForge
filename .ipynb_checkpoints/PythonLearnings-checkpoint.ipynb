# Literals

#### Literal is a raw data given in a variable. In Python, there are various types of literals they are as follows:

#### Numeric Literals 

```python
a = 0b1010 #Binary Literals
b = 100 #Decimal Literal
c = 0o310 #Octal Literal
d = 0x12c #Hexadecimal Literal

#Float Literal
float_1 = 10.5
 #e = 10 to the power  
float_2 = 1.5e2   
float_3 = 1.5e-3

#Complex Literal
x = 3.14j

print(a, b, c, d)
print(float_1, float_2,float_3)
print(x, x.imag, x.real)
```

#### String Literals

```python
string = 'This is Python'
strings = "This is Python"
char = "C"
multiline_str = """This is a multiline string with more than one line code."""
unicode =u"\U0001f600\U0001F606\U0001F923"
raw_str = r"raw \n string"

print(string)
print(strings)
print(char)
print(multiline_str)
print(unicode)
print(raw_str)
```

#### Boolean Literals

```python
a = True + 4 
b = False + 10

print("a:", a)
print("b:", b)
```

#### Special Literals

# Operators

Operators are used to perform operations on variables and values. Python has the following operators:

1. Arithmetic operators
2. Comparison operators
3. Logical operators
4. Bitwise operators
5. Assignment operators
6. Identity operators
7. Membership operators

```python
#Arithmetic 
x,y = 5, 4
print(x+y,x-y)
```

```python
print(x/y ,x%y ,x**y ,x*y)
```

```python
print(x // 2) #integer division
```

```python
#Comparison
print(x>y , x<y)
```

```python
print(x>=y,x<=y,x==y,x!=y)
```

```python
#Logical
x,y = True ,False
print(x and y)
```

```python
print(x or y , not x)
```

```python
#Bitwise works on binary form only
x,y = 2 , 3
print(x & y , x | y)
```

```python
print(x>>2 , y<<2)
```

```python
print(~x)
```

```python
#Assignment : = , += , -= and so on. In python there is no a++ or b-- (Increment/Decrement)
a = 3
```

```python
#Identity - tells that the variable is in the same memory location or not
a,b = 3,3
print(a is b)
```

```python
a,b = [1,2,3],[1,2,3]
print(a is not b)
```

```python
#Membership 
x = 'Delhi'
print("D" not in x)
y = [1,3,2]
print(5 in y)
```

# Decision control statements

```python
# Correct passs and email : satyambhagat200623@gmail.com , Password2006@
email = input("Enter your email")

if "@" in email:
    password = input("Enter the password") 
    
    if email == "satyambhagat200623@gmail.com" and password == "Password2006@":
        print("Welcome")
    
    elif email == "satyambhagat200623@gmail.com" and password != "Password2006@":
        print("Incorrect password")
        password = input("Enter the password") 
        if password == "Password2006@":
            print("Correct Password")
        else: 
            print("Still incorrect")
    
    else:                                                    
        print("Incorrect credentials")
else:            
    print("email is wrong")
```

# Loops 

```python
number = int(input("enter the no : "))
i = 1
while i<11:
    print(number,"x", i ,"=",number * i)
    i += 1
```

```python
import random 
jackpot = random.randint(1,100)
num = int(input("Enter the nuimber"))
counter = 0
while num != jackpot :
    if num < jackpot :
        print("Guess higher")
    else: 
        print("Guess lower")
    num = int(input("Enter the nuimber"))
    counter += 1 
print("Correct guess! You took", counter ,"guess")
```

range has THREE parameters (start, end, step)

```python
list(range(1,11,3)) # step 3 so the difference btw two no.s will be 3
```

```python
# for can iterate any given sequence either it be a list or String
# for i in range(1,10,2):
for i in [1,3,5,7]:
    print(i , end = " ")
```

```python
for i in range(1,6):
    for j in range(0,i):
        print("*", end = " ")
    print(" ")
```

## Break Continue and Pass statements

```python
for i in range (1,11):
    if i == 5:
        break # Terminates the loop 
    print(i)
```

```python
for i in range(1,11):
    if i == 5:
        continue # Skips the code below when the condition is meet
    print(i , end = " ")
```

```python
for i in range (1,11):
    pass # is a no-operation placeholder, ensuring syntactic correctness when no action is needed 
```
